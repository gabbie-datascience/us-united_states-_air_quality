## basic data analysis

```python
# renaming column CO ---> CO2
us_air_quality_copy.rename(columns={'CO':'CO2'}, inplace=True)
```

### basic data tratament

```python
# data tratament - types of dataset, NaN values, descriptive statistics - this is the most recommended
```

### pandas.to_datetime

```python
# this is for if you have a year, month and or day
us_air_quality_copy['Year']=pandas.to_datetime(us_air_quality_copy['Year'], format='%Y')
```

```python
us_air_quality_copy.dtypes.head(1)
```

### .fillna()

```python
# NaN values tratament
```

```python
for column in us_air_quality_copy.columns:
    if us_air_quality_copy[column].isnull().any():
        us_air_quality_copy[column].fillna(us_air_quality_copy[column].mean(), inplace=True)
```

```python
us_air_quality_copy.isna().sum()
```

# states

```python
# important provinces
```

```python
california_state=us_air_quality_copy[us_air_quality_copy['State']=='California']
california_state.head()
```

```python
texas_state=us_air_quality_copy[us_air_quality_copy['State']=='Texas']
texas_state.head()
```

```python
newyork_state=us_air_quality_copy[us_air_quality_copy['State']=='New York']
newyork_state.head()
```

## outliers analysis(exists outliers - there were atypical events)

```python
# imports
```

```python
import matplotlib.pyplot
import seaborn
```

```python
categories = ['PM2.5', 'PM10', 'NO2', 'CO', 'SO2', 'O3', 'VOC', 'Temperature', 'Humidity', 'WindSpeed', 'Pressure', 'Rainfall']
```

```python
# california state
```

```python
fig, axes=matplotlib.pyplot.subplots(3, 4, figsize=(20, 16))
axes=axes.flatten()

for i, category in enumerate(categories):
    axes[i].boxplot(california_state[category])
    axes[i].set_title(category)
    axes[i].set_ylabel('Values')

for j in range(len(categories), len(axes)):
  fig.delaxes(axes[j])

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

```python
# texas state
```

```python
fig, axes=matplotlib.pyplot.subplots(3, 4, figsize=(20, 16))
axes=axes.flatten()

for i, category in enumerate(categories):
    axes[i].boxplot(texas_state[category])
    axes[i].set_title(category)
    axes[i].set_ylabel('Values')

for j in range(len(categories), len(axes)):
  fig.delaxes(axes[j])

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

```python
# new york state
```

```python
fig, axes=matplotlib.pyplot.subplots(3, 4, figsize=(20, 16))
axes=axes.flatten()

for i, category in enumerate(categories):
    axes[i].boxplot(newyork_state[category])
    axes[i].set_title(category)
    axes[i].set_ylabel('Values')

for j in range(len(categories), len(axes)):
  fig.delaxes(axes[j])

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```
